```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
```

```python
from sklearn.datasets import load_iris
data = load_iris()

features = data.data
print np.shape(features)
features_names = data.feature_names
print features_names
target = data.target
target_names = data.target_names
```

```python
fig, axes = plt.subplots(2,3)
pairs = [(0,1), (0,2), (0,3), (1,2), (1,3), (2,3)]

# Set up 3 different pairs of (colour, marker)
colour_markers = [
    ('r','>'),
    ('g','o'),
    ('b','x'),
]
for i, (p0, p1) in enumerate(pairs):
    ax = axes.flat[i]
    for t in range(3):
        c, marker = colour_markers[t]
        ax.scatter(features[target == t, p0], features[
                target == t, p1], marker=marker, c=c)
        ax.set_xlabel(features_names[p0])
        ax.set_ylabel(features_names[p1])
        ax.set_xticks([])
        ax.set_yticks([])

fig.tight_layout()
```

```python
# Get an array for strings using NumPy's indexing
labels = target_names[target]
# The petal length is the feature at position 2
petalLength = features[:, 2]
# Build an array of booleans:
is_setosa = (labels == 'setosa')
max_setosa = petalLength[is_setosa].max()
min_non_setosa = petalLength[~is_setosa].min()
print('Maximum of setosa: {0}.'.format(max_setosa))
print('Minimum of others: {0}.'.format(min_non_setosa))
```

# Simple decision classification model
Therefore using just the petal length, the Iris Setosa flower can be separated from the others (Iris Virginica or Iris Versicolor).

The tough part is to classify the remaining flowers.

```python
# Select only the non-Setosa features and labels
features_non_setosa = features*np.transpose((~is_setosa,~is_setosa,~is_setosa,~is_setosa))

# Create a new target variable
is_virginica = (labels == 'virginica')
```

```python
# Initialize the best_acc to impossibly low value
best_acc = -1.0

for fi in range(features_non_setosa.shape[1]):
    # We are going to test all possible thresholds
    thresh = features_non_setosa[:,fi]
    for t in thresh:
        # Get the vector for feature 'fi'
        feature_i = features_non_setosa[:, fi]
        # apply threshold 't'
        pred = feature_i > t
        acc = (pred == is_virginica).mean()
        rev_acc = (pred == ~is_virginica).mean()
        if rev_acc > acc:
            reverse = True
            acc = rev_acc
        else:
            reverse = False
            
        if acc > best_acc:
            best_acc = acc
            best_fi = fi
            best_t = t
            best_reverse = reverse
```

```python
print(best_fi, best_t, best_reverse, best_acc)

COLOUR_FIGURE = True

t = best_t
t2 = 1.75

# Features to use: 3 & 2
f0, f1 = best_fi, 2

if COLOUR_FIGURE:
    area1c = (1., .8, .8)
    area2c = (.8, .8, 1.)
else:
    area1c = (1., 1, 1)
    area2c = (.7, .7, .7)

# Plot from 90% of smallest value to 110% of largest value
# (all feature values are positive, otherwise this would not work very well)

x0 = features[:, f0].min() * .9
x1 = features[:, f0].max() * 1.1

y0 = features[:, f1].min() * .9
y1 = features[:, f1].max() * 1.1

fig,ax = plt.subplots()
ax.fill_between([t, x1], [y0, y0], [y1, y1], color=area2c)
ax.fill_between([x0, t], [y0, y0], [y1, y1], color=area1c)
ax.plot([t, t], [y0, y1], 'k--', lw=2)
ax.plot([t2, t2], [y0, y1], 'k:', lw=2)
ax.scatter(features[is_virginica, f0],
            features[is_virginica, f1], c='b', marker='o', s=40)
ax.scatter(features[~is_virginica, f0],
            features[~is_virginica, f1], c='r', marker='x', s=40)
ax.set_ylim(y0, y1)
ax.set_xlim(x0, x1)
ax.set_xlabel(features_names[f0])
ax.set_ylabel(features_names[f1])
fig.tight_layout()
```

```python
# Create a function to fit results
def fit_model(features, labels):
    '''Learn a simple threshold model'''
    best_acc = -1.0
    # Loop over all the features:
    for fi in range(features.shape[1]):
        thresh = features[:, fi].copy()
        # test all feature values in order:
        thresh.sort()
        for t in thresh:
            pred = (features[:, fi] > t)

            # Measure the accuracy of this 
            acc = (pred == labels).mean()

            rev_acc = (pred == ~labels).mean()
            if rev_acc > acc:
                acc = rev_acc
                reverse = True
            else:
                reverse = False
            if acc > best_acc:
                best_acc = acc
                best_fi = fi
                best_t = t
                best_reverse = reverse

    # A model is a threshold and an index
    return best_t, best_fi, best_reverse


# This function predicts class
def predict(model, features):
    '''Apply a learned model'''
    # A model is a pair as returned by fit_model
    t, fi, reverse = model
    if reverse:
        return features[:, fi] <= t
    else:
        return features[:, fi] > t

# Measure accuracy of classifier
def accuracy(features, labels, model):
    '''Compute the accuracy of the model'''
    preds = predict(model, features)
    return np.mean(preds == labels)

correct = 0.0

for ei in range(len(features)):
    # select all but the one at position `ei`:
    training = np.ones(len(features), bool)
    training[ei] = False
    testing = ~training
    model = fit_model(features[training], is_virginica[training])
    predict(model, features[testing])
    predictions = predict(model, features[testing])
    correct += np.sum(predictions == is_virginica[testing])
acc = correct/float(len(features))
print('Accuracy: {0:.1%}'.format(acc))
```
