# Concordance (Word Count)

Find the ten most commonly used words in a text file.

```python
with open('great-expectations.txt', encoding='utf-8-sig') as f:
    text = ''.join(f)
```

```python
text[:50]
```

```python
def concordance(text):
    freq = {}
    for word in text.split():
        if word not in freq:
            freq[word] = 0
        freq[word] += 1
    return freq

concordance(text)
```

```python
from re import sub

def concordance(text):
    freq = {}
    for word in text.split():
        word = sub('[^\w]', '', word.lower())
        if word not in freq:
            freq[word] = 0
        freq[word] += 1
    return freq
```

```python
freq = concordance(text)
sorted(freq.items(), key = lambda kv: kv[1], reverse = True)[:10]
```

```python
from re import sub

def concordance(text):
    freq = {}
    for word in text.split():
        word = sub('[^\w]','', word.lower())
        freq[word] = freq.get(word, 0) + 1
    return freq

freq = concordance(text)
sorted(freq.items(), key = lambda kv: kv[1], reverse = True)[:10]
```

```python
from re import sub
from collections import defaultdict

def concordance(text):
    freq = defaultdict(int)
    for word in text.split():
        word = sub('[^\w]', '', word.lower())
        freq[word] += 1
    return freq

freq = concordance(text)
sorted(freq.items(), key = lambda kv: kv[1], reverse = True)[:10]
```

```python
from re import sub
from collections import Counter

def concordance(text):
    return Counter(sub('[^\w]', '', word.lower()) for word in text.split())

freq = concordance(text)
sorted(freq.items(), key=lambda kv: kv[1], reverse=True)[:10]
```
