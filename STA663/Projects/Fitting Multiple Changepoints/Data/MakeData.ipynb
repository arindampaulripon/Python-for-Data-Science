### Simulate Gaussian Time Series with a Single Change Point

This data set is used for profiling and code optimization, unit test and the demonstration example.

```python
### configuration
from __future__ import division
import os
import sys
import glob
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as stats
import scipy.linalg as la

%matplotlib inline
%precision 4
plt.style.use('ggplot')
```

```python
### Simulation
np.random.seed(1234)
n = 150
m = 1 # number of change points
sigma = np.sqrt(3.0)
tau_true = 50
tau_true_rep = np.array([tau_true, n - tau_true])
s_true = np.repeat(range(0, m+1), tau_true_rep)
theta_true = np.array([1., 3.])
y = np.random.normal(theta_true[s_true], sigma, size = n)
np.savetxt("data_test", y)
plt.figure(figsize=(7, 5))
plt.plot(np.arange(n) + 1, y, '-b', linewidth = 2)
plt.xlabel('Time')
plt.ylabel('y_t')
plt.xlim([0, 151])
plt.title('Gaussian Observations')
```

### Simulate Gaussian Time Series with two Single Change Points

This data set is used for the application of the algorithm, both in terms of model fitting and model comparisons.

```python
### Simulation
np.random.seed(1234)
n = 150
m = 2 # number of change points
sigma = np.sqrt(3.0)
tau_true = np.array([50, 100])
tau_true_rep = np.array([tau_true[0], tau_true[1] - tau_true[0], n - tau_true[1]])
s_true = np.repeat(range(0, m+1), tau_true_rep)
theta_true = np.array([1., 3., 5.])
y = np.random.normal(theta_true[s_true], sigma, size = n)
np.savetxt("data_gaussian", y)
plt.figure(figsize=(7, 5))
plt.plot(np.arange(n) + 1, y, '-b', linewidth = 2)
plt.xlabel('Time')
plt.ylabel('y_t')
plt.xlim([0, 151])
plt.title('Gaussian Observations')
```
