Background for Exercises 1 and 2
----

The first 2 exercises are about using Newton's method to find the cube roots of unity - find $z$ such that $z^3 = 1$. From the fundamental theorem of algebra, we know there must be exactly 3 complex roots since this is a degree 3 polynomial.

We start with Euler's fabulosu equation
$$
e^{ix} = \cos x + i \sin x
$$

Raising $e^{ix}$ to the $n$th power where $n$ is an integer, we get from Euler's formula with $nx$ substituting for $x$
$$
(e^{ix})^n = e^{i(nx)} = \cos nx + i \sin nx
$$

Whenever $nx$ is an integer multiple of $2\pi$, we have
$$
\cos nx + i \sin nx = 1
$$

So
$$
e^{2\pi i \frac{k}{n}}
$$
is a root of 1 whenever $k/n = 0, 1, 2, \ldots$.

So the cube roots of unity are $1, e^{2\pi i/3}, e^{4\pi i/3}$. 

While we can do this analytically, the idea is to use Newton's method to find these roots, and in the process, disover some rather perplexing behavior of Newton's method.

**Exercise 1 (10 points)**. Netwon's method for functions of complex variables - stability and basins of attraction.

1. Write a function with the following function signature `newton(z, f, fprime, max_iter=100, tol=1e-6)` where
    - `z` is a starting value (a complex number e.g.  ` 3 + 4j`)
    - `f` is a function of `z`
    - `fprime` is the derivative of  `f`
The function will run until either max_iter is reached or the absolute value of the Newton step is less than tol. In either case, the function should return the number of iterations taken and the final value of `z` as a tuple (`i`, `z`). 

2. Define the function `f` and `fprime` that will result in Newton's method finding the cube roots of 1. Find 3 starting points that will give different roots, and print both the start and end points. 

```python
# your code here
# 1.
def newton(z, f, fprime, max_iter=100, tol=1e-6):
    " The function using Newton's method to find a local extrema. "
    notconverge = 1
    count = 0
    xvals = []
    xvals.append(z)
    while(notconverge == 1 and count < max_iter):
        diffval = - f(xvals[count]) / fprime(xvals[count])
        if abs(diffval) < tol:
            notconverge = 0
        else:
            xvals.append(xvals[count] + diffval)
            count = count + 1
    return (count, xvals[count])

# 2.
def fcube(x):
    "The cube function"
    return x**3 - 1
def fcube_prime(x):
    "The derivative of the fcube function"
    return 3*x**2
# Starting points
z1, z2, z3 = 0.5, complex(-1, 1), complex(-1, -1)
m1 = newton(z = z1, f = fcube, fprime = fcube_prime, max_iter=100, tol=1e-6)[1]
m2 = newton(z = z2, f = fcube, fprime = fcube_prime, max_iter=100, tol=1e-6)[1]
m3 = newton(z = z3, f = fcube, fprime = fcube_prime, max_iter=100, tol=1e-6)[1]
print "The three starting points with their corresponding roots are:"
print "Start =", z1, "Root = ", m1
print "Start =", z2, "Root = ", m2
print "Start =", z3, "Root = ", m3
```

**Exercise 2 (10 points)**. Write the following two plotting functions to see some (pretty) aspects of Newton's algorithm in the complex plane.

1. The first function `plot_newton_iters(f, fprime, n=200, extent=[-1,1,-1,1], cmap='hsv')` calculates and stores the number of iterations taken for convergence (or max_iter) for each point in a 2D array. The 2D array limits are given by `extent` - for example, when `extent = [-1,1,-1,1]` the corners of the plot are `(-i, -i), (1, -i), (1, i), (-1, i)`. There are `n` grid points in both the real and imaginary axes. The argument `cmap` specifies the color map to use - the suggested defaults are fine. Finally plot the image using `plt.imshow` - make sure the axis ticks are correctly scaled. Make a plot for the cube roots of 1.

2. The second function `plot_newton_basins(f, fprime, n=200, extent=[-1,1,-1,1], cmap='jet')` has the same arguments, but this time the grid stores the identity of the root that the starting point converged to. Make a plot for the cube roots of 1 - since there are 3 roots, there should be only 3 colors in the plot.

```python
# your code here
# 1. We continue to use the previous defined functions
def plot_newton_iters(f, fprime, n=200, extent=[-1,1,-1,1], cmap='hsv'):
    "return a plot with convergence iterations for cube roots of 1 in the complex plane"
    xgrid = [extent[0] + i * float(extent[1] - extent[0]) / float(n - 1) for i in range(n)]
    ygrid = [extent[2] + j * float(extent[3] - extent[2]) / float(n - 1) for j in range(n)]
    iters = [newton(z = complex(i, j), f = f, fprime = fprime, max_iter = 100, tol = 1e-6)[0] for j in ygrid
                                                                                              for i in xgrid]
    zvals = np.array(iters).reshape(n,n)
    img = plt.imshow(zvals, cmap = cmap, extent = extent, origin = 'lower') # specifies the lower options to make sure the data allign well
    plt.colorbar(cmap = cmap)

# Visualize this plot
plt.figure(figsize=(8, 8))
plot_newton_iters(f = fcube, fprime = fcube_prime, n = 200, extent = [-1,1,-1,1], cmap = 'hsv')
```

```python
# 2.
def plot_newton_basins(f, fprime, n = 200, extent = [-1,1,-1,1], cmap = 'jet'):
    " return a plot with unique root identity in the complex plane "
    
    def identity(x, tol = 1e-4):
        "Write an identity function to recognize each unique root"
        if abs(x - m1) < tol:
            return 1
        if abs(x - m2) < tol:
            return 2
        else:
            return 3
    
    xgrid = [extent[0] + i * float(extent[1] - extent[0]) / float(n - 1) for i in range(n)]
    ygrid = [extent[2] + j * float(extent[3] - extent[2]) / float(n - 1) for j in range(n)]
    roots = [newton(z = complex(i, j), f = f, fprime = fprime, max_iter = 100, tol = 1e-6)[1] for j in ygrid
                                                                                              for i in xgrid]
    rootid = np.vectorize(identity)(roots)
    zvals = np.array(rootid).reshape(n,n)
    img = plt.imshow(zvals, cmap = cmap, extent = extent, origin = 'lower')
    plt.colorbar(cmap = cmap, ticks = [1, 2, 3])   
    # The tick names: 'm1 = 1', 'm2 = -0.5+0.866j', 'm3 = -0.5-0.866j'

# Now visualize the plot
plt.figure(figsize=(8, 8))
plot_newton_basins(f = fcube, fprime = fcube_prime, n = 200, extent = [-1,1,-1,1], cmap = 'jet')
```

```python
# your code here
from sympy import symbols, diff, hessian, Function, N, exp, solve
from numpy import e
# 1.
x1, x2 = symbols('x1 x2') # create symbols
f = symbols('f', cls = Function)
f = -x1*x2*exp(-(x1**2 + x2**2) / 2) # create the function
fprime_x1 = diff(f, x1)
fprime_x2 = diff(f, x2)
fprime_x1
```

```python
fprime_x2
```

```python
# 2.
H = hessian(f, [x1, x2])
H
```

```python
# 3.
# Find the critical points of f
critical = solve([fprime_x1, fprime_x2], [x1, x2])
print critical
```

```python
# 4.
# Now we characterize the critical points
import scipy.linalg as la

# Compute the hessian and evaluate whether it is positive/negative definite by eigen decomp
H0 = np.array(N(H.subs([(x1,critical[0][0]), (x2,critical[0][1])])))
H1 = np.array(N(H.subs([(x1,critical[1][0]), (x2,critical[1][1])])))
H2 = np.array(N(H.subs([(x1,critical[2][0]), (x2,critical[2][1])])))
H3 = np.array(N(H.subs([(x1,critical[3][0]), (x2,critical[3][1])])))
H4 = np.array(N(H.subs([(x1,critical[4][0]), (x2,critical[4][1])])))

e0, v0 = la.eig(H0)
e1, v1 = la.eig(H1)
e2, v2 = la.eig(H2)
e3, v3 = la.eig(H3)
e4, v4 = la.eig(H4)

# Check their eigenvalues
print e0
print e1
print e2
print e3
print e4

# Characterization
print "We find the following 2 points are local minina (negative definite Hessian with negative eigenvalues)"
print critical[0]
print critical[4]
print "And the following two are local maxima (positive definite Hessian with positive eigenvalues)"
print critical[1]
print critical[3]
print "The other one is a inflection point"
print critical[2]
```

```python
# Constrained Optimization
from scipy.optimize import minimize
def foo(x):
    "The original function"
    x1, x2 = x[0], x[1]
    return -x1*x2*e**(-(x1**2 + x2**2) / 2)

cons = ({'type': 'eq',
         'fun' : lambda x: np.array([2*x[0] + 3*x[1] - 5]),
         'jac' : lambda x: np.array([2, 3])},
        {'type': 'ineq',
         'fun' : lambda x: np.array([10 - x[0]**2 - x[1]**2])})

# Let's try the four different starting point and summarize the findings
start = np.array([2, 2])
print minimize(foo, start, constraints = cons).x
print "We know that [1, 1] is the mimimum under the specified constraints."
```

**Exercise 4 (20 points)**. Find the maximum likelihood function for a logistic regression. Load the file "train.csv" (located in this directory) into python.  This file is data from the survival data from the Titanic (from https://www.kaggle.com/).  You are to use a logistic regression to model survival as a function of gender, age and class (of travel). Find the maximum likelihood estimator of $\beta$ by numerical optimization using stochastic gradient descent as follows:

1. Stochastic gradient descent.  In this method, gradient descent is used essentially by fitting *one data point at a time*.  Recall the usual gradient descent step:

$$\beta_{i+1} = \beta_i - \nabla \ell(\beta_i)$$

where

$$\nabla \ell(\beta_i) = \sum\limits_{j=1}^n \nabla\ell(\beta_i,x_j,y_j)$$

and $\ell$ is the log-likelihood function.  All of the data is used to make the next step toward the optimal $\beta$.  In stochastic gradient descent, only one point at a time is used to determine the next $\beta$:

$$\beta_{i+1} = \beta_i - \alpha\nabla \ell(\beta_i,x_j,y_j)$$

where $\alpha$ is the step size.  For simplicity, we'll take a constant $\alpha=1$.

Implement the following stochastic gradient algorithm:
 
   a. Shuffle data points (i.e. randomly permute the order of the $(x_j,y_j)$
   
   b. Refine beta using the iterative formula above over each data point. 
   
   c. Repeat a and b until convergence is reached.
 
Apply the algorithm to the given data set to find the best-fit logistic regression coefficients.  Do not forget that your optimization should include a tolerance and a limit on the number of iterations.

```python
# your code here
import pandas as pd
# Read in data
data_c = pd.DataFrame.from_csv("train.csv")
data = data_c[['Survived', 'Pclass', 'Sex', 'Age']]
data = pd.DataFrame(data.dropna()) # exclude all the NAs
n, p = data.shape

# create the design matrix
Pclass2 = np.array(data.Pclass == 2) + 0.0
Pclass3 = np.array(data.Pclass == 3) + 0.0
Sex = np.array(data.Sex == "male") + 0.0

# Since the column Age has really large values compared to others, we normalize this variable
# to ensure the stability of the algorithm
m_A, sd_A = np.mean(data.Age), np.std(data.Age, ddof = 1)
Age = (np.array(data.Age) - m_A) / sd_A

# The response vector and the design matrix
y = np.array(data.Survived)       # response vector
X = np.vstack([[np.ones(n), Sex, Age, Pclass2, Pclass3]]).T # design matrix

# Algorithm
np.random.seed(1234) # set seed
beta = np.zeros(p + 1)
curr_beta = np.zeros(p + 1)
alpha = 0.001 # use a small learning rate
tol = 1e-4
notconverge = 1
max_iter = 1000
count = 0

# Note that we need to minimize the negative log-likelihood, so minus the negative gradient becomes a 'plus'.
while(notconverge == 1 and count < max_iter):
    for i in np.random.permutation(range(n)):
        delta = y[i] * X[i,:] * (1 + e**(sum(X[i,:] * beta)))**(-1) - (1 - y[i]) * X[i,:] * (1 + e**(-sum(X[i,:] * beta)))**(-1)
        beta = beta + alpha * delta
    next_beta = beta
    if np.linalg.norm(next_beta - curr_beta) < tol:
        notconverge = 0
    else: 
        curr_beta = next_beta
        beta = next_beta
        count = count + 1
print "Convergence Iterations:", count   

# Scale back the coef of Age and adjust the intercept
next_beta[2] = next_beta[2] / sd_A
next_beta[0] = next_beta[0] - next_beta[2] * m_A / sd_A

# Print the coefficients (very close to the logistic regression fit in Python or R)
print "Intercept:", next_beta[0].round(4)
print "Sex (male):", next_beta[1].round(4)
print "Age:", next_beta[2].round(4)
print "Pclass2:", next_beta[3].round(4)
print "Pclass3:", next_beta[4].round(4)
```
